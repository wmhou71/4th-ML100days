```python
# 載入基本套件
import pandas as pd
import numpy as np

# 如果是分類問題，請使用 DecisionTreeClassifier，若為回歸問題，請使用 DecisionTreeRegressor
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# 讀取訓練與測試資料
data_path = 'data/data-science-london-scikit-learn/'
df_train = pd.read_csv(data_path + 'train.csv', header = None)
df_train_labels = pd.read_csv(data_path + 'trainLabels.csv', header = None)
df_test = pd.read_csv(data_path + 'test.csv', header = None)
df_train.shape
```

```python
df_train.describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_train, df_train_labels, test_size=0.2, random_state=4)
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)

acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
```

```python
clf = RandomForestClassifier(max_depth=6, n_estimators = 2000, max_features='log2')
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)

acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
```

```python
#submit
y_submit = clf.predict(df_test)
submit = pd.DataFrame(columns = ['Id', 'Solution'])
submit['Solution'] = y_submit
submit['Id'] = list(range(1, len(y_submit)+1))                   
submit.to_csv('submit.csv', index= False)
```

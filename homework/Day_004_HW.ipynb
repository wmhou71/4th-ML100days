```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

```python
# 列出資料的大小 (In[4], Hint : shape)  
app_train.info()
app_train.head()
```

```python
#列出所有欄位
type(app_train)
cols = list(app_train.columns)

#print(cols)
count=0
for i in range(len(cols)):
    if count <10:
        print(cols[i])
        count+=1
    else:
        print('\n', cols[i])
        count=0
```
